```python
import numpy as np
import pandas as pd
from preprocess import *
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import Dense, Conv1D, MaxPooling1D, LSTM, Dropout
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
df = pd.read_csv('eth_data.csv')
df = preprocessing(df, int(1))
print(df)
```

```python
# Define the model function
def model(lookback_factor,batch):
    df = pd.read_csv('eth_data.csv')
    df = preprocessing(df, int(lookback_factor))
    look_back = 6 * int(lookback_factor)
    look_ahead = 12
    features = df[['Open', 'High', 'Low', 'Close', 'Volume', 'rsi', 'bb_hband', 'bb_lband', 'ETH_50MA', 'ETH_200MA']].values
    X, y = create_sequences(features, look_back, look_ahead)

    # Scale features and target independently
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()

    # Flatten X for scaling
    n_samples, n_timestamps, n_features = X.shape
    X = X.reshape((n_samples * n_timestamps, n_features))
    X = scaler_X.fit_transform(X)
    X = X.reshape((n_samples, n_timestamps, n_features))

    # Flatten y for scaling
    y = y.reshape(-1, 1)
    y = scaler_y.fit_transform(y)

    # Split into training and test sets
    train_size = int(len(X) * 0.99)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    # Define the CNN-LSTM model with dropout layers
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(MaxPooling1D(pool_size=2))
    model.add(LSTM(75, activation='relu', return_sequences=True))
    model.add(Dropout(0.15))
    model.add(LSTM(50, activation='relu', return_sequences=True))
    model.add(Dropout(0.15))
    model.add(LSTM(50, activation='relu'))
    model.add(Dropout(0.15))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    early_stopping = EarlyStopping(monitor='val_loss', patience=35, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.02, patience=10, min_lr=0.003)

    # Train the model
    model.fit(X_train, y_train, epochs=80, batch_size=batch, callbacks=[early_stopping, reduce_lr])

    # Make predictions
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_predictions = scaler_y.inverse_transform(train_predictions)
    test_predictions = scaler_y.inverse_transform(test_predictions)
    y_test_inv = scaler_y.inverse_transform(y_test)
    y_train_inv = scaler_y.inverse_transform(y_train)
    X_test_inv = X_test

    plt.figure(figsize=(14, 7))
    plt.plot(y_test_inv, label='Actual ETH Price', color='black')
    plt.plot(test_predictions, label='Prediction with lookback x', linestyle='dashed')
    plt.legend()
    plt.title('Actual vs. Predicted ETH Prices (6 hours ahead)')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()
    return test_predictions, y_test_inv,train_predictions,y_train_inv,X_test_inv
# Get predictions from multiple models with different lookback factors
test_preds_1, y_test,train_preds_1,y_train,X_test = model(1.5,16)
print("1")
test_preds_1_5, _ ,train_preds_1_5,_,_= model(1.5,8)
print("1.5")
test_preds_2, _,train_preds_2,_,_= model(1.5,24)
print("2")
test_preds_2_5, _,train_preds_2_5,_,_ = model(1.5,32)
print("2.5")
test_preds_3, _,train_preds_3,_,_ = model(3,8)
print("3")
test_preds_4, _ ,train_preds_4,_,_= model(3,16)
print("4")
test_preds_5, _ , train_preds_5,_,_= model(4,32)
seperate_test = test_preds_1
# Ensure all arrays have the same length for testing
min_length_test = min(len(test_preds_1), len(test_preds_1_5), len(test_preds_2),
                      len(test_preds_2_5), len(test_preds_3), len(test_preds_4),
                      len(test_preds_5), len(y_test))
min_length_train = min(len(train_preds_1), len(train_preds_1_5), len(train_preds_2),
                      len(train_preds_2_5), len(train_preds_3), len(train_preds_4),
                      len(train_preds_5), len(y_test))

# Trim all arrays to the minimum length for testing
test_preds_1 = test_preds_1[:min_length_test]
test_preds_1_5 = test_preds_1_5[:min_length_test]
test_preds_2 = test_preds_2[:min_length_test]
test_preds_2_5 = test_preds_2_5[:min_length_test]
test_preds_3 = test_preds_3[:min_length_test]
test_preds_4 = test_preds_4[:min_length_test]
test_preds_5 = test_preds_5[:min_length_test]
y_test = y_test[:min_length_test]
train_preds_1 = train_preds_1[:min_length_train]
train_preds_1_5 = train_preds_1_5[:min_length_train]
train_preds_2 = train_preds_2[:min_length_train]
train_preds_2_5 = train_preds_2_5[:min_length_train]
train_preds_3 = train_preds_3[:min_length_train]
train_preds_4 = train_preds_4[:min_length_train]
train_preds_5 = train_preds_5[:min_length_train]
```
